```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
df = pd.read_csv('results/logs.csv')
scores = df['averageScore']
x_smooth = np.linspace(0, len(scores) - 1, 300)
spline = make_interp_spline(range(len(scores)), scores, k=3)
y_smooth = spline(x_smooth)
plt.plot(x_smooth, y_smooth)
plt.xlabel('Index')
plt.ylabel('Average Score')
plt.title('Average Score vs. Index')
plt.show()
```
